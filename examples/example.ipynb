```python
import matplotlib.pyplot as plt
from synthetic_controls.plotting import compare_estimators

from synthetic_controls.estimators import StandardEstimator, MBondEstimator, JamesBondEstimator
from synthetic_controls.data.examples import simulated_age_groups, tobacco_us_states
```

```python
panel, observed_causes = simulated_age_groups()

standard_sc = StandardEstimator(panel, treatment_start_time=15)
M_bond_sc = MBondEstimator(panel, treatment_start_time=15, observed_causes=observed_causes)
James_bond_sc = JamesBondEstimator(panel, treatment_start_time=15, observed_causes=observed_causes)


standard_sc.fit("g45")
M_bond_sc.fit("g45")
James_bond_sc.fit("g45");
```

```python
compare_estimators([standard_sc, M_bond_sc, James_bond_sc])
plt.show()
```
